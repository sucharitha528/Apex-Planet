```python
# 1. Import required libraries
import pandas as pd
import numpy as np
import os

# 2. Check working directory and files
print("Current Directory:", os.getcwd())
print("Files in /content:", os.listdir('/content'))

# 3. Load the dataset
df = pd.read_csv('/content/supermarket_sales.csv')
print("\n--- First 5 Rows ---")
display(df.head())

# 4. Data Familiarization
print("\n--- Shape (Rows, Columns) ---")
print(df.shape)

print("\n--- Column Names ---")
print(df.columns)

print("\n--- Dataset Info ---")
df.info()

print("\n--- Statistical Summary ---")
display(df.describe())

# 5. Data Quality Assessment
print("\n--- Missing Values ---")
print(df.isnull().sum())

print("\n--- Duplicate Rows ---")
print(df.duplicated().sum())

print("\n--- Unique Values per Column ---")
print(df.nunique())

# 6. Data Cleaning
df.drop_duplicates(inplace=True)

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Standardize text columns
df['Gender'] = df['Gender'].str.lower()
df['Payment'] = df['Payment'].str.lower()
df['Customer type'] = df['Customer type'].str.lower()

# 7. Data Transformation (Feature Engineering)
df['Month'] = df['Date'].dt.month_name()

df['Revenue_Category'] = pd.cut(
    df['Total'],
    bins=[0, 100, 300, 600],
    labels=['Low', 'Medium', 'High']
)

# 8. Final Validation
print("\n--- Final Dataset Info ---")
df.info()

print("\n--- Final Missing Values ---")
print(df.isnull().sum())

# 9. Save Cleaned Dataset
output_path = '/content/cleaned_supermarket_sales.csv'
df.to_csv(output_path, index=False)

print(f"\n‚úÖ Task-1 Completed Successfully!")
print(f"üìÅ Cleaned dataset saved as: {output_path}")
```
